```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
%load_ext autoreload
%autoreload 2
```

```python
from scipy.optimize import curve_fit

#fitting function 
def func(x,a,b):
    return a*np.exp(b*x)
def func2(x,a,b):
    return a*np.log(x)+b
```

```python
accuracy = [96.025, 96.95, 97.875, 98.425]
splits = [1,2,3,4]
plt.plot(splits, accuracy, 'ro')
popt, pcov = curve_fit(func2, splits, accuracy)
xFit = np.arange(1.0, 5.0, 0.1)
plt.plot(xFit, func2(xFit, *popt), 'r')
plt.xlabel("Training Size Iteration (50th, 25th, 10th, 5th)")
plt.ylabel("Accuracy (Test Data)")
plt.title("Accuracy vs Training Size")
```

```python


run_time = [17.56, 158.32, 2515.6, 21571.96]
splits = [1,2,3,4]
plt.plot(splits, run_time, 'ro')
popt, pcov = curve_fit(func, splits, run_time)
xFit = np.arange(1.0, 5.0, 0.1)
plt.plot(xFit, func(xFit, *popt), 'r')
plt.xlabel("Training Size Iteration (50th, 25th, 10th, 5th)")
plt.ylabel("Run Time (Seconds)")
plt.title("Run time vs Training Size")
```
